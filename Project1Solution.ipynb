```python
import pandas as pd
air_data = pd.read_csv('Project1CIVE202AirQualityDataSet.csv')
air_data = pd.DataFrame(air_data)
import pandas as pd

air_data_grouped = air_data.groupby('sensor.name')
air_data_pm25 = air_data_grouped.agg(max=('pm2.5_atm', 'max'),
                                                     min=('pm2.5_atm', 'min'),
                                                     mean=('pm2.5_atm', 'mean'),
                                                     median=('pm2.5_atm', 'median'))
air_data_pm25
top5_mean_pm25 = (
    air_data_pm25.sort_values(by = 'mean', ascending = False) 
)
top5_mean_pm25.head(5)

```

```python
top5_med_pm25 = (
    air_data_pm25.sort_values(by = 'median', ascending = False) #.sort_values sorts a specified column
)
top5_med_pm25.head(5)
```

```python
air_data_pm10 = air_data_grouped.agg(max=('pm10.0_atm', 'max'),
                                                     min=('pm10.0_atm', 'min'),
                                                     mean=('pm10.0_atm', 'mean'),
                                                     median=('pm10.0_atm', 'median'))
air_data_pm10
top5_mean_pm10 = (
    air_data_pm10.sort_values(by = 'mean', ascending = False) #.sort_values sorts a specified column
)
top5_mean_pm10.head(5)
```

```python
top5_med_pm10 = (
    air_data_pm10.sort_values(by = 'median', ascending = False) #.sort_values sorts a specified column
)
top5_med_pm10.head(5)
```

```python
def assign_temperature(temperature):
    if temperature < 32:
        return 'Below Freezing'
    elif 32 <= temperature <= 50:
        return 'Cool'
    elif 51 <= temperature <= 70:
        return 'Warm'
    else:
        return 'Hot' #this line means that if temperature is >71 it is Hot
  
```

```python
air_data['temperature_level'] = air_data['temperature'].apply(assign_temperature)
air_data.groupby('temperature_level').agg({
    'pm2.5_atm': ['mean'],
    'pm10.0_atm': ['mean'],
    'voc': ['mean']
})
```

```python
def assign_humidity(humidity):
    if humidity < 50:
        return 'Low Humidity'
    elif 50 <= humidity <= 80:
        return 'High Humidity'
    else:
        return 'Very High Humidity'
        #this line means that if temperature is >71 it is Hot
air_data['humidity_level'] = air_data['humidity'].apply(assign_humidity)
air_data.groupby('humidity_level').agg({
    'pm2.5_atm': ['mean'],
    'pm10.0_atm': ['mean'],
    'voc': ['mean']
})
```
